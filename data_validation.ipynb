```python
import os 
from PIL import Image
from collections import Counter

def load_image_ff(folder_path):
    image_paths = []
    images = []

    if not os.path.exists(folder_path):
        raise FileNotFoundError(f"Not Found:{folder_path}")
    
    for file in os.listdir(folder_path):
        ext = os.path.splitext(file)[1].lower()
        if ext in {".jpg",".jpeg",".png"}:
            image_paths.append(os.path.join(folder_path,file))

    print(f"Found Image Paths: {len(image_paths)}")

    for path in image_paths:
        try:
            image = Image.open(path).convert("RGB")
            images.append(image)

        except Exception as e:
            print("Failed to Load", {e})

    print(f"Loaded Images: {len(images)}")
    return images


def flatten_list(nested_list):
    return [item for sublist in nested_list for item in sublist]



```

```python
import matplotlib.pyplot as plt
import random
from PIL import Image
import numpy as np

def show_image_from_nested_variable(nested_variable ,node_index = None ,max_images = 6, figsize = (12, 6)):
    if node_index is None:
        node = random.choice(nested_variable)
    else:
        node = nested_variable[node_index]

    images = random.sample(node, min(max_images, len(node)))
    plt.figure(figsize=figsize)

    for i , img in enumerate(images):
        plt.subplot(1, len(images),i+1)

        if isinstance(img, Image.Image):
            plt.imshow(img)
        else:
            plt.imshow(np.asarray(img))

        plt.axis("off")

    plt.tight_layout()
    plt.show()

       
```

```python
import pandas as pd 

def create_image_nodes(nested_list, dataset_name):
    nodes =[]
    for sublist in nested_list:
        for filename in sublist:
            node = {
                'filename': filename,
                'dataset': dataset_name,
                'loaded': False #initially, image not loaded
            }
            nodes.append(node)
    return nodes
    
train_nodes = create_image_nodes(medical_train, 'Train')
test_nodes = create_image_nodes(medical_test, 'Test')

print(train_nodes)
print(test_nodes)

image_df = pd.DataFrame(train_nodes + test_nodes)
```

Light-Weight Medical CNN

```python
import torch.nn as nn

class MedicalCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3,16, kernel_size= 3, stride = 1, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride = 2),

            nn.Conv2d(16,32, kernel_size = 3, stride = 1, padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size =2, stride = 2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32*56*56, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)

        )

    def forward(self,x):
        x = self.features(x)
        return self.classifier(x)
    
```

```python
num_classes = len(medical_train)
print("num_classes =", num_classes)
```

```python
model = MedicalCNN(num_classes= num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)

print(model)
print(optimizer)
print("Model and Optimizer Initialized")
```
